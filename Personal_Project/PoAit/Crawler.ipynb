```python
import requests
from bs4 import BeautifulSoup
url = 'http://woorimal.net/hangul/hyundai%20poem-menu.htm'
resp = requests.get(url)
html = BeautifulSoup(resp.content,'html.parser')
url_list=[]
for links in html.findAll('a'):
        url_list.append(links.attrs['href'])

train_txt = []

for i in range(len(url_list)):
    newurl='http://woorimal.net/hangul/'+url_list[i]
    req = requests.get(newurl)
    detail_html = BeautifulSoup(req.content,'html.parser')
    title=detail_html.find('b').get_text()
    author=detail_html.findAll('span')[1].get_text().replace(u'\xa0','')
    head='&nbsp;'*3+title+'<br><br>'+'&nbsp;'*30+author+'<br><br>'
    poet = str(detail_html.findAll('td')[3])
    train_txt.append(head+poet)
    print("{}th iterations".format(i))
```

```python
import pandas as pd
df = pd.DataFrame(train_txt)
```

```python
df.to_csv('train.csv',header=False,index=False)
```
