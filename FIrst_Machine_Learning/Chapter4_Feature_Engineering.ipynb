```python
#일변량 통계 -각 특성과 타깃 사이에 중요한 통계적 관계가 있는지 계산.
#분류의 경우 분산분석(ANOVA, analysis of variance)
#다른 특성과 깊게 연관된 특성은 선택 X
#계산이 매우 빠르고 모델만들 필요 X 
#분류에서는 f_classif (defalut) 회귀에서는 f_regression 선택하여 테스트 결과인 p값에 기초하여 특성 제외
#높은 p값 = 타깃과의 연관성 낮음 = 제외  (임계값을 조정)
#SelectKBest -k개의 특성 선택
#SelectPercentile - 지정된 비율만큼 특성 선택
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectPercentile, f_classif
cancer=load_breast_cancer()
#고정 난수 
rng=np.random.RandomState(42)
noise=rng.normal(size=(len(cancer.data),50)) #50개의 특성을 노이즈 특성으로 추가
X_w_noise=np.hstack([cancer.data,noise])

X_train,X_test,y_train,y_test = train_test_split(X_w_noise, 
                                                 cancer.target,random_state=0,test_size=.5)
#f_classif(default) 와 SelectPercentile 사용 특성의 50% 선택
select=SelectPercentile(score_func=f_classif,percentile=50)
select.fit(X_train,y_train)
X_train_selected=select.transform(X_train)
print("X_train.shape:",X_train.shape)
print("X_train_selected.shape:",X_train_selected.shape)
```

```python
select.get_support() #.get_support() 선택된 특성을 불리언 값으로 나타내주는 함수
```

```python
mask=select.get_support() #(,80)
plt.matshow(mask.reshape(1,-1),cmap='gray_r')#(1,80)  , True-black 
plt.xlabel("Feature Num")
plt.yticks([0])
```

```python
from sklearn.linear_model import LogisticRegression
#테스트 데이터 변환
X_test_selected=select.transform(X_test)
lr=LogisticRegression()
lr.fit(X_train,y_train)
print("전체 특성 사용한 점수 :{:.3f}".format(lr.score(X_test,y_test)))
lr.fit(X_train_selected,y_train)
print("선택 특성 사용한 점수 :{:.3f}".format(lr.score(X_test_selected,y_test)))
```

```python
y=citibike.values
#POSIX 시간 1970.1.1 부터 흐른 시간 (sec)
X=citibike.index.astype("int64").values.reshape(-1,1)//10**9  #10**9로 나누어 변환
```

```python
n_train=184
def eval_on_features(features,target,regressor):
    X_train,X_test=features[:n_train],features[n_train:]
    y_train,y_test=target[:n_train],target[n_train:]
    regressor.fit(X_train,y_train)
    print("Test Set R^2:{:.2f}".format(regressor.score(X_test,y_test)))
    y_pred=regressor.predict(X_test)
    y_pred_train=regressor.predict(X_train)
    plt.figure(figsize=(10,3))
    plt.xticks(range(0,len(X),8),xticks_name,rotation=90,ha='left')
    plt.plot(range(n_train),y_train,label='Train')
    plt.plot(range(n_train,len(y_test)+n_train),y_test,'-',label='Test')
    plt.plot(range(n_train),y_pred_train,'--',label="Train Prediction")
    plt.plot(range(n_train,len(y_test)+n_train),y_pred,'--',label='Test Prediction')
    plt.legend(loc=(1.01,0))
    plt.xlabel("Date")
    plt.ylabel("Rent Num")
    
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,random_state=0)
eval_on_features(X,y,regressor)
```

```python
#테스트 세트 예측 값이 상수 , R^2 -0.04 학습 X
#POSIX특성의 값이 훈련세트에 있는 특성 값의 범위 밖에 있어서 트리모델인 랜덤 포레스트는
#특성범위 밖으로 외삽(extrapolation)불가능.
#POSIX시간(시계열) 제외하고 시간, 요일을 특성으로 
```

```python
X_hour=citibike.index.hour.values.reshape(-1,1)
eval_on_features(X_hour,y,regressor)
```
