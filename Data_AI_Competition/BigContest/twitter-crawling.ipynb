```python
import datetime 
days_range = [] 
start = datetime.datetime.strptime("2020-01-20", "%Y-%m-%d")
end = datetime.datetime.strptime("2020-09-03", "%Y-%m-%d") 
date_generated = [start+datetime.timedelta(days=x) for x in range(0, (end-start).days)] 
for date in date_generated: 
    days_range.append(date.strftime("%Y-%m-%d"))
print("===설정된 트윗 수집 기간: {} ~ {}===".format(days_range[0], days_range[-1]))
print("===총 {}일 간의 데이터 수집===".format(len(days_range)))

```

```python
import time 
keyword='코로나'
start = datetime.datetime.strptime("2020-01-20", "%Y-%m-%d")
end = datetime.datetime.strptime("2020-09-03", "%Y-%m-%d") 
tweetCriteria = got.manager.TweetCriteria().setQuerySearch('{}'.format(keyword))\
                                            .setSince(str(start))\
                                            .setUntil(str(end))\
                                            .setMaxTweets(10000) # 모두 수집
print("데이터 수집 시작========") 
start_time = time.time()
tweet = got.manager.TweetManager.getTweets(tweetCriteria)
print("=== 총 트윗 개수 {} ===".format(len(tweet)))

```

```python
from tqdm.notebook import tqdm

tweet_list = []

for index in tqdm(tweet):
    
    # 메타데이터 목록 
    content = index.text
    tweet_date = index.date.strftime("%Y-%m-%d")
    tweet_list.append([content,tweet_date])
```

```python
for i,tw in enumerate(tweet_list):
    tweet_list[i] = [tw[1],tw[0]]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tw_df = pd.DataFrame(tweet_list)
```
