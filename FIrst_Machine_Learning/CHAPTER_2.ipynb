```python
from sklearn.datasets import load_breast_cancer

cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,
                                               stratify=cancer.target,random_state=66)
training_accuracy=[]
test_accuracy=[]
#1~10 n_neighbors
neighbors_settings=range(1,11)
for n_neighbors in neighbors_settings:
    #Creating Model
    clf=KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train,y_train)
    #save accuracy of train set
    training_accuracy.append(clf.score(X_train,y_train))
    # 일반화 정확도 저장
    test_accuracy.append(clf.score(X_test,y_test))
    
plt.plot(neighbors_settings, training_accuracy,label='Train Accuracy')
plt.plot(neighbors_settings, test_accuracy,label='Test Accuracy')
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor

X,y=mglearn.datasets.make_wave(n_samples=40)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

reg=KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train,y_train)
```

```python
fig,axes=plt.subplots(1,3,figsize=(15,4))
#(-3,3,1000 data point)
line=np.linspace(-3,3,1000).reshape(-1,1)
for n_neighbors,ax in zip([1,3,9],axes):
    reg=KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train,y_train)
    ax.plot(line,reg.predict(line))
    ax.plot(X_train,y_train,'^',c=mglearn.cm2(0),markersize=8)
    ax.plot(X_test,y_test,'v',c=mglearn.cm2(1),markersize=8)
    
    ax.set_title("{}-neighbors train score:{:.2f} test score:{:.2f}"
                 .format(n_neighbors,reg.score(X_train,y_train),reg.score(X_test,y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")

for i in range(3):
    axes[i].legend(["Model Prediction","Train data/target","Test data/target"],loc="best")
```

```python
#Linear regression = OLS(ordinary least squares-최소제곱법)
#mean squared error min - 평균제곱오차 최소화 하기 
from sklearn.linear_model import LinearRegression
X,y=mglearn.datasets.make_wave(n_samples=60)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)

lr=LinearRegression().fit(X_train,y_train)
```

```python
#w-weight(가중치) or coefficient(계수) lr객체 coef_ 속성에 저장
#b-offset(편향) or intercept(절편) intercept_ 속성에저장
print('lr.coef_:',lr.coef_)
print('lr.intercept_:',lr.intercept_)
```

```python
#coef_ 속성은 입력 특성에 하나씩 대응되는 numpy배열
lr.coef_.shape
```

```python
print("R^2 of Train Set:{:.2f}".format(lr.score(X_train,y_train)))
print("R^2 of Test Set:{:.2f}".format(lr.score(X_test,y_test)))
```

```python
#train set 과 test set  R^2가 비슷 -> 과소적합 (1차원 데이터셋이라 모델이 단순)
#고차원에서는 선형 모델의 과대적합 우려
X,y=mglearn.datasets.load_extended_boston()
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
lr=LinearRegression().fit(X_train,y_train)
```

```python
print("R^2 of Train set:{:.2f}".format(lr.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(lr.score(X_test,y_test)))
#Overfitting - 과대적합 
#모델 정확도차이 
```

```python
#Ridge Regression - 리지 회귀
#w의 원소를 0에 가깝게 - 모든 특성이 출력에 주는 영향을 최소화 = Regularization(규제)
#과대 적합 방지
#L2 규제
from sklearn.linear_model import Ridge
ridge=Ridge().fit(X_train,y_train)
print("R^2 of Train set:{:.2f}".format(ridge.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(ridge.score(X_test,y_test)))
```

```python
#alpha parameter- default=1.0  단순화정도 0 - linear regression 과 같음
#alpha 높이면 계수가 0에 가까워진다 = 일반화
ridge10=Ridge(alpha=10).fit(X_train,y_train)
print("R^2 of Train set:{:.2f}".format(ridge10.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(ridge10.score(X_test,y_test)))
```

```python
ridge01=Ridge(alpha=0.1).fit(X_train,y_train)
print("R^2 of Train set:{:.2f}".format(ridge01.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(ridge01.score(X_test,y_test)))
```

```python
plt.plot(ridge10.coef_,'^',label="Ridge alpha=10")
plt.plot(ridge.coef_,'s',label="Ridge alpha=1")
plt.plot(ridge01.coef_,'v',label="Ridge alpha=0.1")

plt.plot(lr.coef_,'o',label="Linear Regression")
plt.xlabel("coefficient")
plt.ylabel("value of coef_")
xlims=plt.xlim()  #xlims=(xmin,xmax)
plt.hlines(0,xlims[0],xlims[1]) # draw line (height, start x , end x)
plt.xlim(xlims)
plt.ylim(-25,25)
plt.legend()
```

```python
#Lasso - make coef near 0  
#L1 규제
# 0이되는 계수 존재 - Feature Selection 특성 선택
from sklearn.linear_model import Lasso
lasso=Lasso().fit(X_train,y_train)
print("R^2 of Train set:{:.2f}".format(lasso.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(lasso.score(X_test,y_test)))
print("Num of The Used Features:",np.sum(lasso.coef_ !=0))
#104개중 4개 특성사용
```

```python
 
# alpha defalut=0 , how strong that lasso changes coef_ to 0
#max_iter(반복실행 최대 횟수) 증가 시켜야함
lasso001=Lasso(alpha=0.01,max_iter=100000).fit(X_train,y_train)
print("R^2 of Train set:{:.2f}".format(lasso001.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(lasso001.score(X_test,y_test)))
print("Num of The Used Features:",np.sum(lasso001.coef_ !=0))
```

```python
lasso00001=Lasso(alpha=0.0001,max_iter=100000).fit(X_train,y_train)
print("R^2 of Train set:{:.2f}".format(lasso00001.score(X_train,y_train)))
print("R^2 of Test set:{:.2f}".format(lasso00001.score(X_test,y_test)))
print("Num of The Used Features:",np.sum(lasso00001.coef_ !=0))
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(
    cancer.data,cancer.target,stratify=cancer.target,random_state=42)
logreg=LogisticRegression().fit(X_train,y_train)
print("R^2 of Train set : {:.3f}".format(logreg.score(X_train,y_train)))
print("R^2 of Test set : {:.3f}".format(logreg.score(X_test,y_test)))
#과소적합
```

```python
logreg100=LogisticRegression(C=100).fit(X_train,y_train)
print("R^2 of Train set : {:.3f}".format(logreg100.score(X_train,y_train)))
print("R^2 of Test set : {:.3f}".format(logreg100.score(X_test,y_test)))
```

```python
logreg001=LogisticRegression(C=0.001).fit(X_train,y_train)
print("R^2 of Train set : {:.3f}".format(logreg001.score(X_train,y_train)))
print("R^2 of Test set : {:.3f}".format(logreg001.score(X_test,y_test)))
```

```python
plt.plot(logreg100.coef_.T,'^',label="C=100")
plt.plot(logreg.coef_.T,'o',label="C=1")
plt.plot(logreg001.coef_.T,'v',label="C=0.01")

plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
xlims=plt.xlim()
plt.hlines(0,xlims[0],xlims[1])
plt.xlim(xlims)
plt.ylim(-5,5)
plt.xlabel("Feature")
plt.ylabel("value of coef_")
plt.legend()
```

```python
for C, marker in zip([0.001,1,100],['o','^','v']):
    lr_l1=LogisticRegression(C=C,penalty="l1").fit(X_train,y_train)
    print("C={:.3f} 인 l1 로지스틱 회귀의 훈련 정확도:{:.2f}".format
         (C,lr_l1.score(X_train,y_train)))
    print("C={:.3f} 인 l1 로지스틱 회귀의 테스트 정확도:{:.2f}".format
         (C,lr_l1.score(X_test,y_test)))
    plt.plot(lr_l1.coef_.T,marker,label="C={:.3f}".format(C))
plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
xlims=plt.xlim()
plt.hlines(0,xlims[0],xlims[1])
plt.xlabel("특성")
plt.ylabel("계수 크기")
plt.ylim(-5,5)
plt.legend(loc='best')
```

```python
#트리기반 회귀 모델은 Extrapolation(외삽) -훈련데이터 범위 밖의 포인트 에 대한 예측이 불가능
import os
ram_prices=pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH,"ram_price.csv"))

plt.yticks(fontname="Arial")
plt.semilogy(ram_prices.date , ram_prices.price) #log scale graph
plt.xlabel("years")
plt.ylabel("prices($/Mbyte)")
```

```python
from sklearn.tree import DecisionTreeRegressor
# before2000 - train set
# after2000 - test set
data_train=ram_prices[ram_prices.date<2000]
data_test=ram_prices[ram_prices.date>=2000]
#가격 예측을 위해 날짜 특성만을 이용
X_train=data_train.date[:,np.newaxis]
#데이터와 타깃 사이의 관계 간단히 - LOG SCALE
y_train=np.log(data_train.price)

tree=DecisionTreeRegressor().fit(X_train,y_train)
linear_reg=LinearRegression().fit(X_train,y_train)
#전체기간에 대해 예측
X_all=ram_prices.date[:,np.newaxis]

pred_tree=tree.predict(X_all)
pred_lr=linear_reg.predict(X_all)
#로그스케일 되돌리기
price_tree=np.exp(pred_tree)
price_lr=np.exp(pred_lr)

plt.semilogy(data_train.date,data_train.price, label="Train Data")
plt.semilogy(data_test.date,data_test.price, label="Test Data")
plt.semilogy(ram_prices.date,price_tree, label="Tree Prediction")
plt.semilogy(ram_prices.date,price_lr, label="LinearRegression Prediction")
plt.legend(loc='best')
#tree model perfect prediction train data
#But out of data - can't predict
#시계열 데이터 -Time Seires  일정시간간격 배치된 데이터의 수열
```
