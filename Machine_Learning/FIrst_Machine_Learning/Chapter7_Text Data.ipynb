```python
#tf-idf (term frequency-inverse docnment frequency)
#다른 문서보다 특정 문서에 자주 나타나는 단어에 높은 가중치 (비지도)
#변환후 l2 정규화 ( 문서의 길이에 영향 받지X)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer #텍스트를 입력받아 BOW만들고 tf-idf변환
from sklearn.pipeline import make_pipeline                #CountVectorizer + TfidfTransformer
pipe=make_pipeline(TfidfVectorizer(min_df=5),LogisticRegression())
param_grid={'logisticregression__C':[0.001,0.01,0.1,1,10]}
grid=GridSearchCV(pipe,param_grid,cv=5)
grid.fit(text_train,y_train)  
print("최상의 교차 검증 점수:{:.2f}".format(grid.best_score_))
```

```python
#tf-idf 가 비지도 학습이라 긍정적/ 부정적 리뷰 레이블과 관계를 알수 없다
vectorizer=grid.best_estimator_.named_steps['tfidfvectorizer']
X_train=vectorizer.transform(text_train)
max_value=X_train.max(axis=0).toarray().ravel() #특성별 최댓값
sorted_by_tfidf=max_value.argsort()
feature_names=np.array(vectorizer.get_feature_names())
print("가장 낮은 tfidf 가진 특성:\n",feature_names[sorted_by_tfidf[:20]])
#전체 문서에 걸쳐 매우 많이 나타나거나 , 조금씩만 사용, 매우긴 문서에서만 사용
print("가장 높은 tfidf 가진 특성:\n",feature_names[sorted_by_tfidf[-20:]])
# 구체적인 정보 포함
```

```python
sorted_by_idf=np.argsort(vectorizer.idf_) #idf_ 에 저장 
print("low idf features:\n",feature_names[sorted_by_idf[:100]])
```

```python
#어간 추출(Stemming) 표제어 추출(Lemmatization) 
#Porter 어간 추출기
import spacy
import nltk
```

```python
en_nlp=spacy.load('en_core_web_sm')
stemmer = nltk.stem.PorterStemmer()
def compare_normalization(doc):
    doc_spacy=en_nlp(doc)
    print("표제어:")
    print([token.lemma_ for token in doc_spacy])
    
    print("어간:")
    print([stemmer.stem(token.norm_.lower()) for token in doc_spacy])
```

```python
compare_normalization(u"Our meeting today was worse than yesterday, "
                      "I'm scared of meeting the clients tomorrow.")
#표제어가 더 복잡하지만 좋은 결과를 주로 낸다
```

```python
import re
from sklearn.feature_extraction.text import CountVectorizer
regexp=re.compile('(?u)\\b\\w\\w+\\b') #CountVectorizer 에서 사용되는 정규 표현식
en_nlp=spacy.load('en_core_web_sm',disable=['parser','ner']) #원본 토큰 분할기 저장
old_tokenizer = en_nlp.tokenizer
en_nlp.tokenizer=lambda string: old_tokenizer.tokens_from_list(regexp.findall(string))
#정규 표현식 이용한 토큰분할기로 변경
def custom_tokenizer(document):  #자작 토큰 분할기 생성
    doc_spacy=en_nlp(document)
    return [token.lemma_ for token in doc_spacy]
lemma_vect=CountVectorizer(tokenizer=custom_tokenizer, min_df=5)
```

```python
#토픽 모델링 topic modeling 
#잠재 디리클레 할당(Latent Dirichlet Allocation, LDA)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(max_features=10000,max_df=0.15)
X=vect.fit_transform(text_train)
```

```python
from sklearn.decomposition import LatentDirichletAllocation 
lda=LatentDirichletAllocation(n_components=10,learning_method="batch",max_iter=25,random_state=0)
document_topics=lda.fit_transform(X)
```

```python
print("lda.components_.shape:",lda.components_.shape) # n_topics, n_words
#(10,10000) - 10개의 토픽에 10000개 단어의 중요도
```

```python
sorting=np.argsort(lda.components_,axis=1)[:,::-1] #정렬을 반대로,중요도를 내림차순으로(default 오름차순)
feature_names=np.array(vect.get_feature_names())
mglearn.tools.print_topics(topics=range(10),feature_names=feature_names,sorting=sorting,
                          topics_per_chunk=5,n_words=10)
```

```python
lda100=LatentDirichletAllocation(n_components=100,learning_method="batch",max_iter=25,random_state=0)
document_topics100=lda100.fit_transform(X)
```

```python
topics=np.array([7,8,9,10,1,84,60,43,70,80,90,31,25,4,97])
sorting=np.argsort(lda100.components_,axis=1)[:,::-1]
feature_names=np.array(vect.get_feature_names())
mglearn.tools.print_topics(topics=topics,feature_names=feature_names,sorting=sorting,
                          topics_per_chunk=5,n_words=20)
```

```python
music=np.argsort(document_topics100[:,45])[::-1]#documets_topics 특성 10개 가진 문서들
for i in music[:10]: 
    print(b".".join(text_train[i].split(b".")[:2])+b".\n") #음악 토픽 비중 큰 문서 10개를 2문장씩 출력
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,10))
topic_names=["{:>2}".format(i)+" ".join(words) for i,words in enumerate(feature_names[sorting[:
                                                                                             ,:2]])]
#열을 2개까지만 채택 (단어)
for col in[0,1]:
    start=col*50
    end=(col+1)*50
    ax[col].barh(np.arange(50),np.sum(document_topics100,axis=0)[start:end])
    ax[col].set_yticks(np.arange(50))
    ax[col].invert_yaxis()
    ax[col].set_yticklabels(topic_names[start:end],ha='left',va='top')
    ax[col].set_xlim(0,2000)
    yax=ax[col].get_yaxis()
    yax.set_tick_params(pad=130)  
plt.tight_layout()
```
