```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
```

```python
#선형 모델을 유연하게 만드는법 - 특성끼리 곱하거나 특성을 거듭제곱하여 새로운 특성 추가
from sklearn.datasets import make_blobs
X,y=make_blobs(centers=4,random_state=8)
y=y%2
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
from sklearn.svm import LinearSVC
linear_svm=LinearSVC()
linear_svm.fit(X,y)

mglearn.plots.plot_2d_separator(linear_svm,X)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
X_new=np.hstack([X,X[:,1:]**2])  #X[:,1] =[1,2,3,4,5] X[:,1:]=[[1],[2],[3],[4],[5]]
'''hstack 명령은 행의 수가 같은 두 개 이상의
배열을 옆으로 연결하여 열의 수가 더 많은 배열을 만든다.
연결할 배열은 하나의 리스트에 담아야 한다. '''
from mpl_toolkits.mplot3d import Axes3D,axes3d
figure=plt.figure()
ax=Axes3D(figure,elev=-152,azim=-26) #시점
#y==0 인포인트 부터 
mask=y==0 #Boolean
ax.scatter(X_new[mask,0],X_new[mask,1],X_new[mask,2],c='b',
          cmap=mglearn.cm2,s=60,edgecolor='k',label="y==0")
ax.scatter(X_new[~mask,0],X_new[~mask,1],X_new[~mask,2],c='r',
          cmap=mglearn.cm2,s=60,edgecolor='k',marker='^',label="y==1")
ax.set_xlabel("Feature 0")
ax.set_ylabel("Feature 1")
ax.set_zlabel("Feature 1 **2")
plt.legend()
```

```python
linear_svm_3d=LinearSVC().fit(X_new,y)
coef,intercept=linear_svm_3d.coef_.ravel(),linear_svm_3d.intercept_

figure=plt.figure()
ax=Axes3D(figure,elev=-152,azim=-26)
xx=np.linspace(X_new[:,0].min()-2,X_new[:,0].max()+2,50)
yy=np.linspace(X_new[:,1].min()-2,X_new[:,1].max()+2,50)
XX, YY=np.meshgrid(xx,yy)
ZZ=(coef[0]*XX+coef[1]*YY+intercept)/-coef[2]
ax.plot_surface(XX,YY,ZZ,rstride=8,cstride=8,alpha=0.3)
ax.scatter(X_new[mask,0],X_new[mask,1],X_new[mask,2],c='b',
          cmap=mglearn.cm2,s=60,edgecolor='k',label="y==0")
ax.scatter(X_new[~mask,0],X_new[~mask,1],X_new[~mask,2],c='r',
          cmap=mglearn.cm2,s=60,edgecolor='k',marker='^',label="y==1")
ax.set_xlabel("Feature 0")
ax.set_ylabel("Feature 1")
ax.set_zlabel("Feature 1 **2")
plt.legend()
```

```python
ZZ=YY**2
dec=linear_svm_3d.decision_function(np.c_[XX.ravel(),YY.ravel(),ZZ.ravel()]) #ravel() -1차원배열로!
plt.contourf(XX,YY,dec.reshape(XX.shape),levels=[dec.min(),0,dec.max()], #두 개의 1차원 배열을 칼럼으로 
                              #levels 등고선 표시위치                   #세로로 붙여서 2차원 배열 만들기
           cmap=mglearn.cm2,alpha=0.5)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
#Kernel trick -커널 기법  : 새로운 특성을 만들지 않고 고차원에서 분류기 학습
#확장된 특성에 대한 데이터 포인트들의 거리(스칼라곱)을 계산
#1.다항식커널: 지정된 차수까지 계산(특성1 **2 X 특성2 **5)
#2.가우시안 커널 = RBF(Radial Basis Function)커널-모든 차수의 모든 다항식 고려 
```

```python
#서포트벡터 - 두 클래스 사이 경계에 위치한 데이터 포인트
#새로운 데이터 포인트 예측 - 서포트 벡터와의 거리 측정 (서포트벡터 중요도  dual_coef_)
from sklearn.svm import SVC
X,y=mglearn.tools.make_handcrafted_dataset()
svm=SVC(kernel='rbf',C=10,gamma=0.1).fit(X,y)
mglearn.plots.plot_2d_separator(svm,X,eps=1) #eps-확대?
mglearn.discrete_scatter(X[:,0],X[:,1],y)
sv=svm.support_vectors_ #데이터 포인트들
#dual_coef_의 부호에 의해 서포트 벡터의 클래스 레이블이 결정
sv_labels=svm.dual_coef_.ravel()>0
mglearn.discrete_scatter(sv[:,0],sv[:,1],sv_labels,s=15,markeredgewidth=3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
svm.dual_coef_.ravel()<0
```

```python
#gamma >0 작으면 데이터 포인트의 영향범위가 커진다
#C는 규제 매개변수 (각 포인트의 중요도 dual_coef_ 제한)
fig,axes=plt.subplots(3,3,figsize=(15,10))
for ax,C in zip(axes,[-1,0,3]):
    for a,gamma in zip(ax,range(-1,2)):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma,ax=a)
axes[0,0].legend(["Class 0","Class 1",'Class 0 Support Vector','Class 1 Support Vector'],ncol=4,loc=(.9,1.2))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)
svc=SVC()
#Default [C=1,gamma=1/n * features]
svc.fit(X_train,y_train)
print("Accuracy of Train_Set:{:.2f}%".format(svc.score(X_train,y_train)*100))
print("Accuracy of Test_Set:{:.2f}%".format(svc.score(X_test,y_test)*100))
```

```python
plt.boxplot(X_train)
plt.xticks(np.linspace(0,30,30))
plt.yscale("symlog")
plt.xlabel("Feature")
plt.ylabel("Size of Feature")
```

```python
#특성들의 자릿수 차이가 매우커서 커널SVM에 영향이 아주 크다.
#모든 특성 평균0 단위 분산  or  0~1로 맞추기
min_on_training=X_train.min(axis=0) #훈련 세트에서 특성별 최솟값 계산 axis=0 행방향(높이)  axis=1 열방향 
range_on_training=(X_train-min_on_training).max(axis=0)# 최댓값 - 최솟값 범위 계산
#훈련데이터에 최솟값을 빼고 범위로 나누면 각 특성에 대해 최솟값은 0 최댓값은 1
X_train_scaled=(X_train-min_on_training)/range_on_training
print("특성별 최솟값\n",X_train_scaled.min(axis=0))
print("특성별 최댓값\n",X_train_scaled.max(axis=0))
```

```python
X_test_scaled=(X_test-min_on_training)/range_on_training
```

```python
svc=SVC()
svc.fit(X_train_scaled,y_train)
print("Accuracy of Train_Set:{:.2f}%".format(svc.score(X_train_scaled,y_train)*100))
print("Accuracy of Test_Set:{:.2f}%".format(svc.score(X_test_scaled,y_test)*100))
#UNDERFITTING! (Similiar Accuracy)
```

```python
svc=SVC(C=1000)
svc.fit(X_train_scaled,y_train)
print("Accuracy of Train_Set:{:.2f}%".format(svc.score(X_train_scaled,y_train)*100))
print("Accuracy of Test_Set:{:.2f}%".format(svc.score(X_test_scaled,y_test)*100))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
X,y=make_moons(n_samples=100,noise=0.25,random_state=3)
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=42)

mlp=MLPClassifier(solver='lbfgs',random_state=0).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.6)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
#기본 비선형 함수는 Rectified Linear Unit
```

```python
mlp=MLPClassifier(solver='lbfgs',random_state=0,hidden_layer_sizes=[10]).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.6)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
#은닉 유닛이 10개
```

```python
mlp=MLPClassifier(solver='lbfgs',random_state=0,hidden_layer_sizes=[10,10]).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.6)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
#10개의 유닛으로된 2개의 은닉층 
```

```python
mlp=MLPClassifier(solver='lbfgs',random_state=0,hidden_layer_sizes=[10,10],
                  activation='tanh').fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.6)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
#tanh 활성화 함수가 적용
```

```python
fig,axes=plt.subplots(2,4,figsize=(20,8))
for axx,n_hidden_nodes in zip(axes,[10,100]):
    for ax,alpha in zip(axx,[0.0001,0.001,0.1,1]):
        mlp=MLPClassifier(solver='lbfgs',random_state=0,
                          hidden_layer_sizes=[n_hidden_nodes,n_hidden_nodes],alpha=alpha)
        mlp.fit(X_train,y_train)
        mglearn.plots.plot_2d_separator(mlp,X_train,alpha=.3,fill=True,ax=ax)
        mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,ax=ax)
        ax.set_title("n_hidden=[{},{}]\nalpha={:.4f}".format(n_hidden_nodes,n_hidden_nodes,
                                                            alpha))
        
```

```python
#신경망도 모든 입력특성을 평균0 분산1 이되도록 하는것이 좋다
mean_on_train=X_train.mean(axis=0) #평균      (각특성)
std_on_train=X_train.std(axis=0)  #표준 편차
#데이터에서 평균을 빼고 표준 편차로 나누면 
#평균 0 표준 편차 1인 데이터로 변환   -  IMPORTANT
X_train_scaled=(X_train-mean_on_train)/std_on_train
X_test_scaled=(X_test-mean_on_train)/std_on_train

mlp=MLPClassifier(random_state=0)
mlp.fit(X_train_scaled,y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles
X,y=make_circles(noise=0.25,factor=0.5,random_state=1)
y_named=np.array(["blue","red"])[y]  # 0 > blue  1> red

X_train,X_test,y_train_named,y_test_named,y_train,y_test=train_test_split(X,y_named,y,random_state=0)
gbrt=GradientBoostingClassifier(random_state=0).fit(X_train,y_train_named)
```

```python
#이진 분류에서 decision_fuction 반환값의 크기는 (n_samples,)이며 각 샘플이 하나의 실수값을 반환
print("X_test.shape:",X_test.shape)
print("결정 함수 결과 형태:",gbrt.decision_function(X_test).shape)
```

```python
#이 값은 모델이 데이터포인트가 양성 클래스인 클래스 1에 속한다고 믿는 정도.(음수는 다른 클래스)
print("결정 함수:\n",gbrt.decision_function(X_test)[:6])
```

```python
print("임계치와 결정 함수 결과 비교:\n",gbrt.decision_function(X_test)>0)
print("예측:\n",gbrt.predict(X_test))
```

```python
gbrt.classes_  #이진 분류에서 음성 클래스는 classes_속성의 첫 번째 원소
```

```python
greater_zero=(gbrt.decision_function(X_test)>0).astype(int) #불리언 값을 0,1 변경 
pred=gbrt.classes_[greater_zero]
print("예측결과:",np.all(pred==gbrt.predict(X_test)))
```

```python
decision_function=gbrt.decision_function(X_test)
print("결정 함수 최솟값: {:.2f} 최댓값: {:.2f}".format(np.min(decision_function),np.max(decision_function)))
```

```python
fig,axes=plt.subplots(1,2,figsize=(13,5))
mglearn.tools.plot_2d_separator(gbrt,X,ax=axes[0],alpha=.4,
                               fill=True,cm=mglearn.cm2)
scores_image=mglearn.tools.plot_2d_scores(gbrt,X,ax=axes[1],alpha=.4,cm=mglearn.ReBl)
for ax in axes:
    mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,ax=ax,markers='^')
    mglearn.discrete_scatter(X_test[:,0],X_test[:,1],y_test,ax=ax,markers='o')
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
cbar=plt.colorbar(scores_image,ax=axes.tolist())
cbar.set_alpha(1)
cbar.draw_all()
axes[0].legend(["Test Class 0","Test Class 1","Train Class 0","Train Class 1"],ncol=4,loc=(.1,1.1))
```

```python
print("확률 값의 형태:",gbrt.predict_proba(X_test).shape)
#이진 분류에서 (n_samples,2)
#[0] - 첫 번째 클래스의 예측 확률 [1] -2번째
#0~1 사이값 - 확률  두 클래스 확률합=1
```

```python
print("예측 확률:\n",gbrt.predict_proba(X_test)[:6])
#복잡도가 낮으면 예측 불확실성이 높다
```
